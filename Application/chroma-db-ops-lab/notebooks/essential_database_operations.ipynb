# Essential Database Operations in Chroma DB

This notebook provides a hands-on guide to essential database operations in Chroma DB, including creating collections, adding documents, and performing similarity searches.

```python
# Import necessary libraries
from chroma_client import ChromaClient
from operations.collections import create_collection, list_collections
from operations.documents import add_document, update_document, get_document
from operations.similarity_search import perform_similarity_search

# Initialize Chroma DB client
client = ChromaClient()

# Create a collection for employee data
collection_name = 'employees'
create_collection(collection_name)

# List collections to verify creation
print("Collections:", list_collections())
```

## Adding Documents

Next, we will add documents to the `employees` collection using data from the `employees.json` file.

```python
# Load employee data from JSON file
import json

with open('../data/employees.json') as f:
    employees = json.load(f)

# Add each employee document to the collection
for employee in employees:
    add_document(collection_name, employee)
print("Documents added to collection.")
```

## Modifying Documents

We can also update existing documents in the collection. Here, we will update an employee's information.

```python
# Update an employee document
employee_id = '123'  # Example employee ID
updated_info = {'name': 'John Doe', 'position': 'Senior Developer'}
update_document(collection_name, employee_id, updated_info)
print("Document updated.")
```

## Performing Similarity Searches

Finally, we will perform a similarity search based on a query.

```python
# Perform a similarity search
query = 'Software Engineer'
results = perform_similarity_search(collection_name, query)
print("Similar results:", results)
```
